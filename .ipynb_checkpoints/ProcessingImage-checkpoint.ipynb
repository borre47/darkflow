```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
options = {
    'model':'cfg/yolov2-tiny-7c.cfg',
    'load':3689,    
    'threshold' : 0.1
}
#'load':'bin/yolov2.weights'
tfnet = TFNet(options)
```

```python
img = cv2.imread('DataSet/train/images/img0ga.jpg')
result =tfnet.return_predict(img)
result
```

```python
def draw_bb(img, result):
    for r in result:
        xmin_ymin = (r['topleft']['x'],r['topleft']['y'])
        xmax_ymax = (r['bottomright']['x'],r['bottomright']['y'])
        label = r['label'] +' '+ str(round(r['confidence']*100,2)) +'%'
        img = cv2.rectangle(img,xmin_ymin,xmax_ymax,(0,255,0),7)
        img = cv2.putText(img,label,xmin_ymin,cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),2)
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        
```

```python
img_bb=draw_bb(img,result)
plt.figure( figsize=[10,10])
plt.imshow(img_bb)
plt.show()
```

## Predict on a image folder and save the out images

```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt
from pascal_voc_writer import Writer
%config InlineBackend.figure_format = 'svg'
options = {
    'model':'cfg/yolov2-tiny-7c.cfg',
    'load':3929,    
    'threshold' : 0.3
}


#2039

#6250
#9290
#10136
#'load':'bin/yolov2.weights'
tfnet = TFNet(options)
```

```python
def draw_bb(img, result):
    for r in result:
        xmin_ymin = (r['topleft']['x'],r['topleft']['y'])
        xmax_ymax = (r['bottomright']['x'],r['bottomright']['y'])
        label = r['label'] +' '+ str(round(r['confidence']*100,2)) +'%'
        img = cv2.rectangle(img,xmin_ymin,xmax_ymax,(0,255,0),3)
        img = cv2.putText(img,label,xmin_ymin,cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),1)
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        
```

```python
from os import listdir
test_path = 'DataSet/test/images/'
out_path = 'DataSet/out/'
imageslist = listdir(test_path)
predictions = [] 
for image in imageslist:
    #if 'shelve' not in image: continue 
    img = cv2.imread(test_path + image,cv2.IMREAD_COLOR)
    height,width,_ = img.shape
   
    result =tfnet.return_predict(img)
    predictions.append(result)
    img_bb=draw_bb(img,result)
    cv2.imwrite(out_path+'images/'+image[:-4]+'out.jpg', img)
    
    writer = Writer(out_path+'images/'+image[:-4]+'out.jpg', width,height) # imge name    
    for r in result:
        xmin,ymin = r['topleft']['x'],r['topleft']['y']
        xmax,ymax = r['bottomright']['x'],r['bottomright']['y']
        label = r['label'] 
        writer.addObject(label,xmin,ymin,xmax,ymax) # label,xmin,y_min,x_max,y_max    
    writer.save(out_path+'labels/'+image[:-4]+'out.xml')
```
